```python
import pandas as pd
import numpy as np
```

```python
reactions = pd.read_csv('Reactions.csv')
content = pd.read_csv('Content.csv')
reaction_types = pd.read_csv('ReactionTypes.csv')
```

# Data cleaning ( reactions ) table

```python
reactions
```

#Change coloumn name from 'Type' to 'Reaction Type'

```python
reactions.rename(columns={'Type': 'Reaction Type'}, inplace=True)
reactions
```

### Drop irrelevent coloumn (User ID)

```python
reactions.drop(columns=['User ID'], inplace=True)
reactions
```

### Drop any rows with missing values

```python
reactions.dropna(inplace=True)
reactions
```

```python
reactions.info()
```

# Data cleaning ( content ) table

```python
content
```

### Remove irrelevent coloumns from Content

```python
content.drop(columns=['User ID', 'URL'], inplace=True)
content
```

### replaces double quotes in the Category column


```python
content['Category'] = content['Category'].str.replace('"', '')
```

# Rename the 'Type' column to 'Content Type'

```python
content.rename(columns={'Type': 'Content Type'}, inplace=True)
content
```

# reaction_types table

```python
reaction_types
```

```python
reaction_types.rename(columns={'Type': 'Reaction Type'}, inplace=True)
reaction_types
```

```python
# Merge reactions and content tables on 'Content ID'
merged_data = pd.merge(reactions, content, on='Content ID', how='left')
merged_data
```

```python
final_data = pd.merge(merged_data, reaction_types, on='Reaction Type', how='left')

final_data
```

# Groups the data by 'Category' and calculates the sum of 'Score' for each category.

```python
final_data.groupby('Category')['Score'].sum()
```

```python
# Group the dataframe by 'Category' and sum the 'Score' for each category
category_scores = final_data.groupby('Category')['Score'].sum()

# Sort the categories by their total scores in descending order and select the top 5
top_5_categories = category_scores.sort_values(ascending=False).head(5)

print(top_5_categories)
```

```python
top5_category = final_data['Category'].value_counts().nlargest(5).index.tolist() # find the top 5 catagory
final_data['Top 5 Category'] = final_data['Category'].apply(lambda x: x if x in top5_category else 'Other') # add top 5 catagory column

final_data
```

# Exports the dataframe to a csv file called 'final_data.csv'

```python
final_data.to_csv('final_data.csv')
```
